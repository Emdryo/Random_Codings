```python
def sandpile_grid(N,amount = 1):
    return np.random.randint(0,4,size = (amount,N,N))
```

```python

def improved(N, amount):
    min_val = N**2*2-1
    grids = np.random.randint(0, 4, size=(amount, N, N))
    sums = np.sum(grids, axis=(1, 2))
    mask = sums < min_val

    replacement_count = np.sum(mask)
    while replacement_count > 0:
        new_grids = np.random.randint(0, 4, size=(replacement_count, N, N))
        grids[mask] = new_grids
        sums[mask] = np.sum(new_grids, axis=(1, 2))
        mask = sums < min_val
        replacement_count = np.sum(mask)

    return grids
```

```python
def sandpile_color(N, Grid):
    cmap = colors.ListedColormap(['white','lightgray','gray','black','red', 'blue'])
    bounds = [0, 1, 2, 3, 4, 5,6]
    norm = colors.BoundaryNorm(bounds, cmap.N)

    fig, ax = plt.subplots()
    ax.imshow(Grid, cmap=cmap, norm=norm, origin='lower')
    ax.set_xticks([])
    ax.set_yticks([])
    # Add colorbar
    cbar = plt.colorbar(ax.imshow(Grid, cmap=cmap, norm=norm, origin='lower'), ax=ax, ticks=range(6))
    cbar.ax.set_yticklabels(['0', '1', '2', '3', '4', '>=5'])
    plt.show()
```

```python
def ordinary(x,y,N,Grid, color): #Ordinary boundaries
    Grid[x,y] -= 4
    if y == N-1 and x == N-1: #Upper right corner
        Grid[y-1,x] += 1
        Grid[y,x-1] += 1

    elif x == 0 and y == N - 1: #Upper left
        Grid[1,N-1] += 1  
        Grid[0,N-2] += 1 

    elif x == 0 and y == 0: #Bottom left corner
        Grid[1,0] += 1
        Grid[0,1] += 1

    elif x == N-1 and y == 0: #Bottom right corner
        Grid[N-1,1] += 1
        Grid[N-2,0] += 1

    elif x == 0: #Left side
        Grid[1,y]   += 1
        Grid[0,y+1] += 1
        Grid[0,y-1] += 1   

    elif x == N-1: #Right side
        Grid[N-2,y]   += 1
        Grid[N-1,y+1] += 1
        Grid[N-1,y-1] += 1

    elif y == 0: #Bottom 
        Grid[x,1]   += 1
        Grid[x+1,0] += 1
        Grid[x-1,0] += 1

    elif y == N-1: #Right side
        Grid[x,N-2]   += 1
        Grid[x+1,N-1] += 1
        Grid[x-1,N-1] += 1

    else: 
        Grid[x,y-1] += 1
        Grid[x,y+1] += 1
        Grid[x+1,y] += 1
        Grid[x-1,y] += 1
    if color == True:
        sandpile_color(N,Grid)
    return Grid
```

```python
def periodic(x,y,N,Grid, color): #Periodic boundaries
    Grid[x,y] -= 4
    if y == N-1 and x == N-1: #Upper right corner
        Grid[y-1,x] += 1
        Grid[y,x-1] += 1
        Grid[0,N-1] += 1
        Grid[N-1,0] += 1 

    elif x == 0 and y == N - 1: #Upper left
        Grid[0, 0] += 1 
        Grid[1,N-1] += 1  
        Grid[N-1, N-1] += 1
        Grid[0,N-2] += 1 

    elif x == 0 and y == 0: #Bottom left corner
        Grid[1,0] += 1
        Grid[0,1] += 1
        Grid[N-1,0] += 1
        Grid[0,N-1] += 1

    elif x == N-1 and y == 0: #Bottom right corner
        Grid[N-1,1] += 1
        Grid[N-1,N-1] += 1
        Grid[N-2,0] += 1
        Grid[0,0] += 1

    elif x == 0: #Left side
        Grid[1,y]   += 1
        Grid[0,y+1] += 1
        Grid[0,y-1] += 1
        Grid[N-1,y] += 1
        
    elif x == N-1: #Right side
        Grid[N-2,y]   += 1
        Grid[0,y]     += 1
        Grid[N-1,y+1] += 1
        Grid[N-1,y-1] += 1

    elif y == 0: #Bottom 
        Grid[x,1]   += 1
        Grid[x+1,0] += 1
        Grid[x-1,0] += 1
        Grid[x,N-1] += 1

    elif y == N-1: #Right side
        Grid[x,N-2]   += 1
        Grid[x,0]     += 1
        Grid[x+1,N-1] += 1
        Grid[x-1,N-1] += 1

    else: 
        Grid[x,y-1] += 1
        Grid[x,y+1] += 1
        Grid[x+1,y] += 1
        Grid[x-1,y] += 1
    if color == True:
        sandpile_color(N,Grid)
    return Grid
```

```python
def sandpile_model(Grid,N, boundaries, problem_solver = True): #General model
    x,y = np.random.randint(0,N),np.random.randint(0,N) #Makes a random index
    if problem_solver == True:
        Grid = Grid[0]
    sandpile_color(N,Grid)
    Grid[y,x] += 1 #Updates the chosen index
    sandpile_color(N,Grid) #Plots
    if Grid[y, x] >= 4: #The rule
        Event = True #Just to keep on going
        Grid = boundaries(y, x, N, Grid, True)
        while Event:
            index_x, index_y = np.where(Grid >= 4)
            if len(index_x) == 0:  # No cells to update
                Event = False
            else:
                for i, j in zip(index_x, index_y):
                    Grid = boundaries(i, j, N, Grid, True)
    return Grid
data = [
       [5, 4, 1],
       [4, 5, 0],
       [1, 0, 4]
]

matrix = np.array(data)
print(np.sum(matrix))
#matrix = sandpile_grid(10)
sandpile_color(3,matrix)
#Grid = sandpile_model(matrix,10,periodic, True)
```

# File functions

```python
def save_grid_state(Grid, file_path): #Saving the grids to a file
    with open(file_path, 'a') as file:
        np.savetxt(file, Grid, fmt='%d')
        file.write('\n')  # Add an empty line after each grid, helps
```

```python
def parse_saved_grids(file_path): #Turns a grid-file into a list of grids
    saved_grids = []
    with open(file_path, 'r') as file:
        grid = []
        for line in file:
            if line.strip():  # Non-empty line
                row = list(map(int, line.split()))
                grid.append(row)
            else:  # Empty line indicates the end of a grid
                saved_grids.append(np.array(grid))
                grid = []
    return saved_grids
```

```python
def clear_file(file_path):
    open(file_path, 'w').close()  # Clears it
```

```python
def count_grids_in_file(file_path): #Just for data
    count = 0
    with open(file_path, 'r') as file:
        for line in file:
            if line.strip() == "":
                count += 1
    return count
```

```python
def check_duplicate(saved_grids, new_grid): #To figure out if it is recurirng or not
    for grid in saved_grids:
        if np.array_equal(grid, new_grid):
            return True
    return False
```

```python
def sandpile_save(Grid, N, boundaries, file_path,problem_solver = True): #To find different recurring solutions 
    grid_path = f"Grid_states_{N}x{N}.txt"
    recurrence_path = f"Recurrence_{N}x{N}.txt"
    if problem_solver == True:
        Grid = Grid[0]
    if not os.path.exists(grid_path):
        with open(grid_path, 'w'): pass
    if not os.path.exists(recurrence_path):
        with open(recurrence_path, 'w'): pass

    x, y = np.random.randint(0, N), np.random.randint(0, N)
    Grid[y, x] += 1
    Grid0 = np.copy(Grid)
    old_solutions = parse_saved_grids(grid_path)

    if check_duplicate(old_solutions,Grid): #We don't want to save old solutions
        return #print("Duplicate - Run again")
    save_grid_state(Grid0,grid_path) # Saves the first grid to the grid_list for all previous tried grids
    save_grid_state(Grid, file_path)  # Save the initial grid state for the temporary file
    
    if Grid[y, x] >= 4:
        Event = True
        Grid = boundaries(y, x, N, Grid, False)
        save_grid_state(Grid, file_path)    #Update the saved grid
        while Event:
            index_x, index_y = np.where(Grid >= 4)
            if len(index_x) == 0:  # No cells to update
                Event = False
            else:
                for i, j in zip(index_x, index_y): #Loops over all index x and index y where the value is equal or larger than 4. 
                    Grid = boundaries(i, j, N, Grid, False)
                    Grid_iterations = parse_saved_grids(file_path) #All previous values
                    if check_duplicate(Grid_iterations,Grid): #To see if it is present in the current iteration eg. a duplicate/recurrence
                        save_grid_state(Grid0,recurrence_path)
                        #print("Periodic solution found!")
                        return clear_file("temp_grid.txt")
                    save_grid_state(Grid, file_path)  # Save the updated grid state to check if it duplicates
    return clear_file("temp_grid.txt")

```

```python
def sandpile_matrix(Grid, N, boundaries = periodic,old_grids = [],recurrence_matrix = [], iteration_length = [], size = True): #To find different recurring solutions 
    i_length = 0 #To find the length of each iteration
    temp_iteration = [] 

    if np.max(Grid) < 3:# or any((Grid == old_grid).all() for old_grid in old_grids): #
        return old_grids,recurrence_matrix,iteration_length    #Not interested if it isn't possible for it to topple
    
    if np.sum(Grid) < N*N*2-1: #Not interested if it isn't possible for it to topple
        return old_grids,recurrence_matrix,iteration_length
    
    if size == False: #Problems with the sandpile_grid, this solves it. Turn "False" if you don't use it. 
        Grid = Grid[0]

    xlist, ylist = np.where(Grid == 3)
    x, y = np.random.choice(xlist), np.random.choice(ylist)
    #x, y = np.random.randint(0, N), np.random.randint(0, N)
    Grid[y, x] += 1
    Grid_original = np.copy(Grid)
    old_grids.append(np.copy(Grid))
    temp_iteration.append(np.copy(Grid))
    if Grid[y, x] >= 4:
        Event = True
        Grid = boundaries(y, x, N, Grid, False)
        i_length += 1
        temp_iteration.append(np.copy(Grid))  #same as with sandpile_save
        while Event:
            index_x, index_y = np.where(Grid >= 4)
            if len(index_x) == 0:  # No cells to update
                Event = False
            else:
                for i, j in zip(index_x, index_y): #Loops over all index x and index y where the value is equal or larger than 4. 
                    Grid = boundaries(i, j, N, Grid, False)
                    i_length += 1
                    if any((Grid == iteration).all() for iteration in temp_iteration): #To see if it is present in the current iteration eg. a duplicate/recurrence
                        recurrence_matrix.append(Grid_original)
                        iteration_length.append(i_length)
                        return old_grids,recurrence_matrix, iteration_length
                    temp_iteration.append(np.copy(Grid))# Save the updated grid state to check if it duplicates
    return old_grids,recurrence_matrix, iteration_length
```

```python
def recursion_loop(N,gridsize, removal = True, old_grids = [], recursion_grids = [],it_length = [], boundary = periodic):
    grids = sandpile_grid(N,gridsize)
    old_removed = []
    recursion_removed = []
    iteration_lengths = []
    
    for i in grids:
        old_grids, recursion_grids,it_length = sandpile_matrix(i,N,boundary,old_grids,recursion_grids, it_length)

    if removal:
        for grid1 in old_grids:
            grid_tuple = tuple(grid1.flatten()) #For grid states
            if grid_tuple not in old_removed:
                old_removed.append(grid_tuple)

        for grid2,len_id in zip(recursion_grids,it_length):
            grid1_tuple = tuple(grid2.flatten())
            if grid1_tuple not in recursion_removed:
                recursion_removed.append(grid1_tuple) #For recursions
                iteration_lengths.append(len_id)
        return old_removed, recursion_removed, iteration_lengths
    else: 
        return old_grids, recursion_grids, it_length
```

```python
def grid_sum(old_grid,recurrence):
    sum_old_grid =  np.sum(old_grid, axis = (1,2)) 
    sum_recurrence = np.sum(recurrence, axis = (1,2))

    # sum_old_grid = [np.sum(grid) for grid in old_grid]
    # sum_recurrence = [np.sum(grid) for grid in recurrence]
    return sum_old_grid, sum_recurrence
```

```python
def iteration_plot(iterations, log = True):
    widthnew = np.max(iterations)-np.min(iterations)
    if log:
        plt.hist(iterations, bins = widthnew, label = f"Log histogram for {N}^2 Grid. Minimum value ={np.min(iterations)} ")
        plt.xscale('log')
        plt.yscale('log')
    else:
        if widthnew == 1:
            plt.hist(iterations, label = f"Histogram for {N}^2 Grid. Minimum value = {np.min(iterations)} ")
        else:
            plt.hist(iterations, bins = widthnew, label = f"Iteration histogram for {N}^2 Grid. Minimum value ={np.min(iterations)} ")
    plt.xlabel("Length of iteration before repetition")
    plt.ylabel("Recursions/Repetition value")
    plt.legend()
    plt.show()
```
