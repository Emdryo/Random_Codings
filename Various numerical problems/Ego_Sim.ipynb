```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Rounds = 20
range = [-1,0,1]
Ego_Number = np.random.choice(range, Rounds) # -1, = sur, 0 = meh, 1 =  glad
bet1 = np.random.choice(range, Rounds)
bet2 = np.random.choice(range, Rounds)
bet3 = np.random.choice(range, Rounds)
bet1_in,bet2_in,bet3_in = 10,10,10
bet1list,bet2list,bet3list = [bet1_in],[bet2_in],[bet3_in]
for j,i in enumerate(Ego_Number):
    if bet1_in > 0:
        if bet1[j] == i:
            bet1_in += 1
        else:
            bet1_in -= 1
    else:   bet1_in = bet1_in
    if bet2_in > 0:
        if bet2[j] == i:
            bet2_in += 2
        else:
            bet2_in -= 2
    else:   bet2_in = bet2_in

    if bet3_in > 0:
        if bet3[j] == i:
            bet3_in += 3
        else:
            bet3_in -= 3
    else:   bet3_in = bet3_in
    if bet1_in < 0: bet1_in = 0
    if bet2_in < 0: bet2_in = 0
    if bet3_in < 0: bet3_in = 0
    bet1list.append(bet1_in)
    bet2list.append(bet2_in)
    bet3list.append(bet3_in)
    if bet1_in and bet2_in and bet3_in == 0:
        break
plt.plot(bet1list, label = 'betting 1 per round ')
plt.plot(bet2list, label = 'betting 2 per round ')
plt.plot(bet3list, label = 'betting 3 per round ')
plt.xlabel('Rounds')
plt.ylabel('Ego cash')
plt.legend()
plt.show()
```
