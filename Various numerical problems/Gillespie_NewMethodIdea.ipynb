```python
import numpy as np
import matplotlib.pyplot as plt
r = np.random
import numba as nb
from numba import njit
```

```python

S = 100
I = 10
R = 0
alpha = S**-1.7
beta = 2*S**-1.2
States = np.array([S, I, R])  
History = np.array([States])
Matrix = np.zeros((3,2))
@njit
def SIRMatrix(Matrix,States):
    S,I,R = States
    Matrix[0,0] = -1
    Matrix[1,0] = 1
    Matrix[1,1] = -1
    Matrix[2,1] = 1
def SIRrun(States,alpha,beta,plot = False):
    t = 0
    S,I,R = States
    History = np.array([States])
    Matrix = np.zeros((3,2))
    SIRMatrix(Matrix,States)
    times = [t]
    while (R < (S+I+R)): #or t > 2000:
        S, I, R = States
        rates = [alpha*S*I, beta*I]
        Ratesum = np.sum(rates)
        if Ratesum == 0:
            break
        dt = r.exponential(1/(Ratesum))
        t += dt
        times.append(t)
        j = np.searchsorted(np.cumsum(rates), np.random.rand()*Ratesum)
        States = States +Matrix[:,j]
        History = np.vstack((History, States))
    if plot:
        plt.plot(History[:,0], label="S")
        plt.plot(History[:,1], label="I")
        plt.plot(History[:,2], label="R")
        plt.legend()
        print(len(History[:,2]))
        plt.show()
    return History
```
