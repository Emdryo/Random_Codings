```python
import numpy as np
import matplotlib.pyplot as plt

elcykler = np.array([11,7,18,14,15,17,24,13,27,22,4,12,18,20,22,4,19])
cykler = np.array([92,116,129,160,129,174,252,78,205,157,42,94,68,105,107,45,202])
total_cykler = cykler+elcykler
ratio = elcykler/total_cykler
plt.hist(ratio, bins = int(np.sqrt(len(cykler))), histtype='step')
print(f" Der er i gennemsnittet {ratio.mean()*100:.2f} +- {ratio.std()*1/np.sqrt(len(total_cykler))*100:.2f}% elcykler på gaden")
plt.show()
# Test om de er konsistente

```
