```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
first=1
second=0
third=0
fourth=0
t=0
list1=[]
cash=0
N=20
Limit=[10,100]
while t<N:
    cash+=first#+second+third
    first+=second
    second+=third
    list1=np.append(cash,list1)   
    if cash>=Limit[0]:
        second+=1
        cash=cash-Limit[0]
        Limit[0]=int(np.exp(2)*Limit[0]-Limit[0]*2)
    if cash>=Limit[1]:
        third+=1
        cash=cash-Limit[1]
        Limit[1]=np.exp(3)*Limit[1]-Limit[1]
    t+=1
list1.sort()
#print(list1)
x=np.linspace(0,N,N)
plt.plot(x,list1)
plt.xlabel('Iteration')
plt.ylabel('Cash')
plt.title('Cash in each iteration')
```
